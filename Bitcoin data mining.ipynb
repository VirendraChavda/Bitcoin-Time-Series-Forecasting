```python
# Start from time stamp 1717455600 as now
now = 1724300000
limit = 2000
change = 3600 * limit
timestamp = now - change

btc = []
for i in range(1, 10):
    api_url = f'https://min-api.cryptocompare.com/data/v2/histohour?fsym=BTC&tsym=USD&limit=2000&toTs={timestamp}&e=CCCAGG'
    data = requests.get(api_url)
    data = data.json()
    btc.append(data)
    timestamp = timestamp - change
    #print(api_url)
    
eth = []
timestamp = now - change
for i in range(1,10):
    api_url = f'https://min-api.cryptocompare.com/data/v2/histohour?fsym=ETH&tsym=USD&limit=2000&toTs={timestamp}&e=CCCAGG'
    data = requests.get(api_url)
    data = data.json()
    eth.append(data)
    timestamp = timestamp - change
    
bnb = []
timestamp = now - change
for i in range(1,10):
    api_url = f'https://min-api.cryptocompare.com/data/v2/histohour?fsym=BNB&tsym=USD&limit=2000&toTs={timestamp}&e=CCCAGG'
    data = requests.get(api_url)
    data = data.json()
    bnb.append(data)
    timestamp = timestamp - change

usdt = []
timestamp = now - change
for i in range(1,10):
    api_url = f'https://min-api.cryptocompare.com/data/v2/histohour?fsym=USDT&tsym=USD&limit=2000&toTs={timestamp}&e=CCCAGG'
    data = requests.get(api_url)
    data = data.json()
    usdt.append(data)
    timestamp = timestamp - change

sol = []
timestamp = now - change
for i in range(1,10):
    api_url = f'https://min-api.cryptocompare.com/data/v2/histohour?fsym=SOL&tsym=USD&limit=2000&toTs={timestamp}&e=CCCAGG'
    data = requests.get(api_url)
    data = data.json()
    sol.append(data)
    timestamp = timestamp - change
```

```python
btc_time = []
btc_high = []
btc_low = []
btc_open = []
btc_volumefrom = []
btc_volumeto = []
btc_close = []

for i in range(len(btc)):
    for j in range(len(btc[i]["Data"]["Data"])):
        btc_time.append(btc[i]["Data"]["Data"][j]["time"])
        btc_high.append(btc[i]["Data"]["Data"][j]["high"])
        btc_low.append(btc[i]["Data"]["Data"][j]["low"])
        btc_open.append(btc[i]["Data"]["Data"][j]["open"])
        btc_volumefrom.append(btc[i]["Data"]["Data"][j]["volumefrom"])
        btc_volumeto.append(btc[i]["Data"]["Data"][j]["volumeto"])
        btc_close.append(btc[i]["Data"]["Data"][j]["close"])
        
btc1_time = []
btc1_volume = []

for i in range(len(btc_)):
    for j in range(len(btc_[i]["Data"])):
        btc1_time.append(btc_[i]["Data"][j]["time"])
        btc1_volume.append(btc_[i]["Data"][j]["volume"])
        
btc2_time = []
btc2_top_tier_volume_quote = []
btc2_top_tier_volume_base = []
btc2_top_tier_volume_total = []
btc2_cccagg_volume_quote = []
btc2_cccagg_volume_base = []
btc2_cccagg_volume_total = []
btc2_total_volume_quote = []
btc2_total_volume_base = []
btc2_total_volume_total = []

for i in range(len(btc__)):
    for j in range(len(btc__[i]["Data"])):
        btc2_time.append(btc__[i]["Data"][j]["time"])
        btc2_top_tier_volume_quote.append(btc__[i]["Data"][j]["top_tier_volume_quote"])
        btc2_top_tier_volume_base.append(btc__[i]["Data"][j]["top_tier_volume_base"])
        btc2_top_tier_volume_total.append(btc__[i]["Data"][j]["top_tier_volume_total"])
        btc2_cccagg_volume_quote.append(btc__[i]["Data"][j]["cccagg_volume_quote"])
        btc2_cccagg_volume_base.append(btc__[i]["Data"][j]["cccagg_volume_base"])
        btc2_cccagg_volume_total.append(btc__[i]["Data"][j]["cccagg_volume_total"])
        btc2_total_volume_quote.append(btc__[i]["Data"][j]["total_volume_quote"])
        btc2_total_volume_base.append(btc__[i]["Data"][j]["total_volume_base"])
        btc2_total_volume_total.append(btc__[i]["Data"][j]["total_volume_total"])
        
eth_time = []
eth_open = []
eth_close = []

for i in range(len(eth)):
    for j in range(len(eth[i]["Data"]["Data"])):
        eth_time.append(eth[i]["Data"]["Data"][j]["time"])
        eth_open.append(eth[i]["Data"]["Data"][j]["open"])
        eth_close.append(eth[i]["Data"]["Data"][j]["close"])
        
bnb_time = []
bnb_open = []
bnb_close = []

for i in range(len(bnb)):
    for j in range(len(bnb[i]["Data"]["Data"])):
        bnb_time.append(bnb[i]["Data"]["Data"][j]["time"])
        bnb_open.append(bnb[i]["Data"]["Data"][j]["open"])
        bnb_close.append(bnb[i]["Data"]["Data"][j]["close"])
        
usdt_time = []
usdt_open = []
usdt_close = []

for i in range(len(usdt)):
    for j in range(len(usdt[i]["Data"]["Data"])):
        usdt_time.append(usdt[i]["Data"]["Data"][j]["time"])
        usdt_open.append(usdt[i]["Data"]["Data"][j]["open"])
        usdt_close.append(usdt[i]["Data"]["Data"][j]["close"])
        
sol_time = []
sol_open = []
sol_close = []

for i in range(len(sol)):
    for j in range(len(sol[i]["Data"]["Data"])):
        sol_time.append(sol[i]["Data"]["Data"][j]["time"])
        sol_open.append(sol[i]["Data"]["Data"][j]["open"])
        sol_close.append(sol[i]["Data"]["Data"][j]["close"])
```

```python
btc_df = pd.DataFrame(
    {'datetime': btc_time,
     'high': btc_high,
     'low': btc_low,
     'open': btc_open,
     'volumefrom': btc_volumefrom,
     'volumeto': btc_volumeto,
     'close': btc_close
    })

btc1_df = pd.DataFrame(
    {'datetime': btc1_time,
     'hourly_volume': btc1_volume
    })

btc2_df = pd.DataFrame(
    {'datetime': btc2_time,
     'top_tier_volume_quote': btc2_top_tier_volume_quote,
     'top_tier_volume_base': btc2_top_tier_volume_base,
     'top_tier_volume_total': btc2_top_tier_volume_total,
     'cccagg_volume_quote': btc2_cccagg_volume_quote,
     'cccagg_volume_base': btc2_cccagg_volume_base,
     'cccagg_volume_total': btc2_cccagg_volume_total,
     'total_volume_quote': btc2_total_volume_quote,
     'total_volume_base': btc2_total_volume_base,
     'total_volume_total': btc2_total_volume_total
    })

eth_df = pd.DataFrame(
    {'datetime': eth_time,
     'open': eth_open,
     'close': eth_close
    })

bnb_df = pd.DataFrame(
    {'datetime': bnb_time,
     'open': bnb_open,
     'close': bnb_close
    })

usdt_df = pd.DataFrame(
    {'datetime': usdt_time,
     'open': usdt_open,
     'close': usdt_close
    })

sol_df = pd.DataFrame(
    {'datetime': sol_time,
     'open': sol_open,
     'close': sol_close
    })
```

```python
sol_df.duplicated().sum()
```

```python
df_list = [btc_df, btc1_df, btc2_df, eth_df, bnb_df, usdt_df, sol_df]
for each in df_list:
    each.drop_duplicates(inplace=True)
    each.reset_index(drop = True, inplace = True)
```

```python
final = btc_df.copy()
final = pd.merge(final, btc1_df, how = "left", on = "datetime")
final = pd.merge(final, btc2_df, how = "left", on = "datetime")
```

```python
final["eth_open"] = eth_df["open"]
final["eth_close"] = eth_df["close"]
final["bnb_open"] = bnb_df["open"]
final["bnb_close"] = bnb_df["close"]
final["usdt_open"] = usdt_df["open"]
final["usdt_close"] = usdt_df["close"]
final["sol_open"] = sol_df["open"]
final["sol_close"] = sol_df["close"]
```

```python
final['datetime'] = pd.to_datetime(final['datetime'], unit='s')
```

```python
final.sort_values(by = ['datetime'], inplace = True)
final.reset_index(drop = True, inplace = True)
```
