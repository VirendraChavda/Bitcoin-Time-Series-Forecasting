```python
bit_df = pd.read_csv('Data/bitcoin.csv')
bit_df.head()
```

```python
bit_df['hourly_return'] = 0
bit_df['3_hourly_return'] = 0
bit_df['5_hourly_return'] = 0
bit_df['12_hourly_return'] = 0

for i in range(1, len(bit_df.index)):
    bit_df['hourly_return'][i] = bit_df['close'][i] - bit_df['close'][i-1]

for i in range(3, len(bit_df.index)):
    bit_df['3_hourly_return'][i] = bit_df['close'][i] - bit_df['close'][i-3]
    
for i in range(5, len(bit_df.index)):
    bit_df['5_hourly_return'][i] = bit_df['close'][i] - bit_df['close'][i-5]
    
for i in range(12, len(bit_df.index)):
    bit_df['12_hourly_return'][i] = bit_df['close'][i] - bit_df['close'][i-12]
```

```python
#bit_df['%change'] = ((bit_df['close'] - bit_df['open'])/bit_df['open'])*100
bit_df['3_hourly_volatility'] = bit_df["hourly_return"].rolling(3).std()
bit_df['5_hourly_volatility'] = bit_df["hourly_return"].rolling(5).std()
bit_df['12_hourly_volatility'] = bit_df["hourly_return"].rolling(12).std()
bit_df['daily_volatility'] = bit_df["hourly_return"].rolling(24).std()
```

```python
bit_df = bit_df[(bit_df['datetime'] >= '2022-07-01') & (bit_df['datetime'] < '2023-07-01')]
bit_df.sort_values('datetime', inplace = True)
bit_df.reset_index(drop = True, inplace = True)
```

```python
bit_df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
bit_df.head()
```

```python
trend_df = pd.read_csv('Data/trends.csv')
trend_df.head()
```

```python
trend_df = trend_df[(trend_df['date'] >= '2022-07-01') & (trend_df['date'] < '2023-07-01')]
trend_df.sort_values('date', inplace = True)
trend_df.reset_index(drop = True, inplace = True)
trend_df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
trend_df.rename(columns={"date": "datetime"}, inplace = True)
trend_df.head()
```

```python
sentiment_df = pd.read_csv('Data/reddit_sentiment.csv')
sentiment_df.head()
```

```python
sentiment_df.drop('Unnamed: 0', axis = 1, inplace = True)
sentiment_df.rename(columns={"Datetime": "datetime"}, inplace = True)
```

```python
bit_df['datetime'] = pd.to_datetime(bit_df['datetime'], infer_datetime_format=True)
trend_df['datetime'] = pd.to_datetime(trend_df['datetime'], infer_datetime_format=True)
sentiment_df['datetime'] = pd.to_datetime(sentiment_df['datetime'], infer_datetime_format=True)
```

```python
trend_aggregate = trend_df.groupby('datetime',  as_index=False).sum()
```

```python
trend_aggregate
```

```python
trend_list = []
for i in range(len(bit_df.index)):
    if bit_df['datetime'][i] in trend_aggregate['datetime'].unique():
        row = trend_aggregate[trend_aggregate['datetime'] == bit_df['datetime'][i]]
        row.reset_index(inplace = True)
        trend_list.append(row['btc/bitcoin'][0])
    else:
        trend_list.append(np.nan)
```

```python
bit_df['trend_score'] = trend_list
```

```python
sentiment_aggregate = sentiment_df.groupby('datetime',  as_index=False).sum()
```

```python
pos_list = []
neg_list = []
comp_list = []
pol_list = []
for i in range(len(bit_df.index)):
    if bit_df['datetime'][i] in sentiment_aggregate['datetime'].unique():
        row = sentiment_aggregate[sentiment_aggregate['datetime'] == bit_df['datetime'][i]]
        row.reset_index(inplace = True)
        pos_list.append(row['Pos_score'][0])
        neg_list.append(row['Neg_score'][0])
        comp_list.append(row['Compound_score'][0])
        pol_list.append(row['Polarity'][0])
    else:
        pos_list.append(np.nan)
        neg_list.append(np.nan)
        comp_list.append(np.nan)
        pol_list.append(np.nan)
```

```python
bit_df['Pos_score'] = pos_list
bit_df['Neg_score'] = neg_list
bit_df['Compound_score'] = comp_list
bit_df['Polarity'] = pol_list
```

### Dealing with null values

```python
bit_df.isnull().sum()
```

```python
bit_df['trend_score'].fillna((bit_df['trend_score'].mean()), inplace=True)
bit_df['Pos_score'].fillna((bit_df['Pos_score'].mean()), inplace=True)
bit_df['Neg_score'].fillna((bit_df['Neg_score'].mean()), inplace=True)
bit_df['Compound_score'].fillna((bit_df['Compound_score'].mean()), inplace=True)
bit_df['Polarity'].fillna(0, inplace=True)
```

```python
bit_df.head()
```

```python
bit_df['time_of_the_day'] = np.nan
for i in range(len(bit_df.index)):
    bit_df['time_of_the_day'][i] = bit_df['datetime'][i].hour
```

```python
bit_df['day'] = bit_df['datetime'].dt.day_name()
bit_df['day_of_month'] = bit_df['datetime'].dt.day
```

```python
bit_df['weekend'] = 0
for i in range(len(bit_df.index)):
    if bit_df['day'][i] == 'Sunday' or bit_df['day'][i] == 'Saturday':
        bit_df['weekend'][i] = 1
```

```python
#bit_df['time_index'] = bit_df['datetime'].dt.time
bit_df.head()
```

```python
bit_df.to_csv("Data/bitcoin without lag.csv", index = False)
```
